```python
capitals = pd.read_csv("archive/capital_cities.csv")

capitals.head()
```

```python
capital_data = pd.DataFrame()

capital_data['Country'] = ['Australia', 'Brazil', 'United Kingdom', 'Germany', 'France', 'Sweden',
             'Spain', 'Norway', 'Scotland','Russia','Mexico','South Korea']
```

```python
merged_data = pd.merge(capital_data, capitals, how = 'left', on = "Country")

merged_data.iloc[8, 1] = "Edinburgh"

merged_data
```

```python
#I need to use the google maps api to call and get the lat & long of each city 
#add a location column with merged values to use in API call 
merged_data['Location'] = merged_data['Capital City'] + ", " + merged_data['Country']

merged_data
```

```python
#run a loop through the dataframe to call each city 

latitude = []
longitude = []

for index, row in merged_data.iterrows():
    target_city = row['Location']
    
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_city, g_key)
    
    response = requests.get(target_url)
    
    data = response.json()
    
    lat = data["results"][0]["geometry"]["location"]["lat"]
    lng = data["results"][0]["geometry"]["location"]["lng"]

    latitude.append(lat)
    longitude.append(lng)
    
merged_data['Lat'] = latitude
merged_data['Lng'] = longitude

merged_data
```

```python
genres = ('australian alternative pop' , 'brazilian techno' , "british children's music" , 'german underground rap' , 'french tech house',
'swedish classical', 'spanish jazz', 'norwegian hip hop' , 'scottish electronic' , 'russian dance' , 'mexican tech house' , 'korean trap')

merged_data['Genre'] = genres

final_merged = pd.merge(merged_data, best_genres, how = 'left', on = "Genre")

final_merged
```

```python
#plot on a map using danceability as the weight layer
# Configure gmaps
gmaps.configure(api_key=g_key)

fig = gmaps.figure()

locations = final_merged[["Lat", "Lng"]]

rating = final_merged["Danceability"].astype(float)

heat_layer = gmaps.heatmap_layer(locations, weights=rating,
                                 dissipating=False, max_intensity=1,
                                 point_radius=10)
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
