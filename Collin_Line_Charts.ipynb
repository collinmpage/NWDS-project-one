# Cleaning & Reorganizing Data

```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read in csv file
data = pd.read_csv("archive/data.csv")
data
```

```python
#checking for outliers for song counts per year
year_count_df = pd.DataFrame(data.value_counts('year'))

year_count_df.sort_index(ascending = False).head(50)
```

```python
#Read in csv file
data_by_year = pd.read_csv("archive/data_by_year.csv")
data_by_year
```

```python
#setting variables

x_axis = data_by_year['year']
Danceability = data_by_year['danceability']
Acousticness = data_by_year['acousticness']
Energy = data_by_year['energy']
Instrumentalness = data_by_year['instrumentalness']
Loudness = data_by_year['loudness']
Tempo = data_by_year['tempo']
```

```python
#Danceability Over Time Chart

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.title('Danceability Over Time' ,fontsize =25)
plt.plot(x_axis, Danceability ,linewidth=2, markersize=10,marker="o",color="steelblue")
plt.xlabel('Year',fontsize =14)
plt.ylabel('Danceability',fontsize =14)

plt.savefig('Line_Chart_Images/Danceability_Solo')
plt.show()
```

```python
#Acousticness Over Time

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.title('Acousticness Over Time' ,fontsize =25)
plt.plot(x_axis, Acousticness ,linewidth=2, markersize=10,marker="o",color="steelblue")
plt.xlabel('Year',fontsize =14)
plt.ylabel('Acousticness',fontsize =14)

plt.savefig('Line_Chart_Images/Acousticness_Solo')
plt.show()
```

```python
#Energy Over Time

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.title('Energy Over Time' ,fontsize =25)
plt.plot(x_axis, Energy ,linewidth=2, markersize=10,marker="o",color="steelblue")
plt.xlabel('Year',fontsize =14)
plt.ylabel('Energy',fontsize =14)

plt.savefig('Line_Chart_Images/Energy_Solo')
plt.show()
```

```python
#Instrumentalness Over Time

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.title('Instrumentalness Over Time' ,fontsize =25)
plt.plot(x_axis, Instrumentalness ,linewidth=2, markersize=10,marker="o",color="steelblue")
plt.xlabel('Year',fontsize =14)
plt.ylabel('Instrumentalness',fontsize =14)


plt.savefig('Line_Chart_Images/Instrumentalness_Solo')
plt.show()
```

```python
#Loudness Over Time

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.title('Loudness Over Time' ,fontsize =25)
plt.plot(x_axis, Loudness ,linewidth=2, markersize=10,marker="o",color="steelblue")
plt.xlabel('Year',fontsize =14)
plt.ylabel('Loudness',fontsize =14)

plt.savefig('Line_Chart_Images/Loudness_Solo')
plt.show()
```

```python
#Tempo Over Time

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.title('Tempo Over Time' ,fontsize =25)
plt.plot(x_axis, Tempo ,linewidth=2, markersize=10,marker="o",color="steelblue")
plt.xlabel('Year',fontsize =14)
plt.ylabel('Tempo',fontsize =14)

plt.savefig('Line_Chart_Images/Tempo_Solo')
plt.show()
```

```python
#Danceability Vs. Acousticness

fig = plt.figure(figsize = (15,10))
plt.title('Danceability Vs. Acousticness', fontsize = 25)
ax1 = fig.add_subplot(111)
ax1.plot(x_axis, Danceability, linewidth = 4, color = 'steelblue')
ax1.set_ylabel('Danceability', color = 'steelblue', fontsize = 20)

ax2 = ax1.twinx()
ax2.plot(x_axis, Acousticness, linewidth = 3, color = 'coral')
ax2.set_ylabel('Acousticness', color='coral', fontsize = 20)
for tl in ax2.get_yticklabels():
    tl.set_color('coral')
    
plt.savefig('Line_Chart_Images/Danceability_vs_Acousticness')
```

```python
#Danceability Vs. Energy

fig = plt.figure(figsize = (15,10))
plt.title('Danceability Vs. Energy', fontsize = 25)
ax1 = fig.add_subplot(111)
ax1.plot(x_axis, Danceability, linewidth = 4, color = 'steelblue')
ax1.set_ylabel('Danceability', color = 'steelblue', fontsize = 20)

ax2 = ax1.twinx()
ax2.plot(x_axis, Energy, linewidth = 3, color = 'coral')
ax2.set_ylabel('Energy', color='coral', fontsize = 20)
for tl in ax2.get_yticklabels():
    tl.set_color('coral')
    
    
plt.savefig('Line_Chart_Images/Danceability_vs_Energy')
```
